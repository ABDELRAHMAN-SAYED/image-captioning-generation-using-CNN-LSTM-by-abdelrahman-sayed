```python
import re
#from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

def preprocess_text(document):
  #عادي انت ممكن تستخدم اي فانكشنز تانية تنظف بيها التكست
        # Remove all the special characters
        document = re.sub(r'\W', ' ', str(document))

        # remove all single characters
        document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)

        # Remove single characters from the start
        document = re.sub(r'\^[a-zA-Z]\s+', ' ', document)

        # Substituting multiple spaces with single space
        document = re.sub(r'\s+', ' ', document)

        # Removing prefixed 'b'
        #document = re.sub(r'^b\s+', '', document)

        # Converting to Lowercase
        document = document.lower()

        # Lemmatization
        tokens = document.split()
        tokens = [stemmer.lemmatize(word) for word in tokens]
        tokens = [word for word in tokens if word not in en_stop]
        tokens = [word for word in tokens if len(word) > 3]

        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
```

```python
sent = preprocess_text("Artificial intelligence, b is the most advanced technology of the present era")
print(sent)
```

```python
def the_corpus_fed_into_model(concatenated_data):
  final_corpus = [preprocess_text(sentence) for sentence in concatenated_data if sentence.strip() !='']

  word_punctuation_tokenizer = nltk.WordPunctTokenizer()
  word_tokenized_corpus = [word_punctuation_tokenizer.tokenize(sent) for sent in final_corpus]
  return word_tokenized_corpus
```

```python
word_tokenized_corpus=the_corpus_fed_into_model(concatenated_data)
```

```python
def rewmove_empty_sentences(word_tokenized_corpus):
  word_tokenized_corpus_without_empty_sentences = [x for x in word_tokenized_corpus if x != []]
  return word_tokenized_corpus_without_empty_sentences
```

```python
word_tokenized_corpus=rewmove_empty_sentences(word_tokenized_corpus)
```

```python
def tokenized_docs_each_alone(train):
  total_word_tokenized_corpus=[]
  for i in range(len(train)):
    article= sent_tokenize(train[i])
    final_corpus = [preprocess_text(sentence) for sentence in article if sentence.strip() !='']
    word_punctuation_tokenizer = nltk.WordPunctTokenizer()
    word_tokenized_corpus = [word_punctuation_tokenizer.tokenize(sent) for sent in final_corpus]
    total_word_tokenized_corpus.append(word_tokenized_corpus)
  return total_word_tokenized_corpus
```

```python
total_word_tokenized_corpus=tokenized_docs_each_alone(train)
```

```python
embedding_size = 60
window_size = 5
min_word = 3
down_sampling = 0.01
```

```python
%%time
ft_model = FastText(word_tokenized_corpus,
                      vector_size=embedding_size,
                      window=window_size,
                      min_count=min_word,
                      sample=down_sampling,
                      sg=1,
                      epochs=100)
```

```python
def final_vectors(total_word_tokenized_corpus):
  final_vectors=[]
  doc_vector=[]
  for i in total_word_tokenized_corpus:
    for j in i:
      for k in j:
        doc_vector.append(ft_model.wv[k])
    res = [sum(t) for t in zip(*doc_vector)]
    finalres=[e/len(doc_vector) for e in res]
    final_vectors.append(finalres)
    doc_vector.clear()
  #final_vectors=array(final_vectors)
  return final_vectors

      
    
```

```python
def tfinal_vectors(total_word_tokenized_corpus):
  final_vectors=[]
  doc_vector=[]
  for i in total_word_tokenized_corpus:
    for j in i:
      for k in j:
        doc_vector.append(ft_model.wv[k])
    res = [sum(t) for t in zip(*doc_vector)]
    finalres=[e/len(doc_vector) for e in res]
    final_vectors.append(finalres)
    doc_vector.clear()
  final_vectors=array(final_vectors)
  return final_vectors

      
    
```

```python
def cos_similarity(test_vector,train_vector):
  result = dot(test_vector, train_vector)/(norm(test_vector)*norm(train_vector))
  return result
```

```python
def search_sentence(sentence):
  test=[]
  test.append(sentence)
  test_word_tokenized_corpus=the_corpus_fed_into_model(test)
  test_word_tokenized_corpus=rewmove_empty_sentences(test_word_tokenized_corpus)
  test_total_word_tokenized_corpus=tokenized_docs_each_alone(test)
  test_final_vectors=tfinal_vectors(test_total_word_tokenized_corpus)
  similarity_dictionary=[]
  for i in test_final_vectors:
    for j in final_vectors:
      res=cos_similarity(i,j)
      similarity_dictionary.append(res)
  test_sentence_with_all_train_docs_similarity=similarity_dictionary
  sorted_similarity_from_test_sentence_and_all_train_docs=np.argsort(test_sentence_with_all_train_docs_similarity)[::-1]
  return sorted_similarity_from_test_sentence_and_all_train_docs,sorted_similarity_from_test_sentence_and_all_train_docs[0],similarity_dictionary[sorted_similarity_from_test_sentence_and_all_train_docs[0]],similarity_dictionary
  
    

```

```python
sorted_docs_descendingly,the_most_similar_doc,similarity,similarity_dictionary=search_sentence('diving in red sea')
print('sorted docs descendingly for the input search sentence: ',sorted_docs_descendingly)
print('the most similar doc for the input search sentence: ',the_most_similar_doc)
print('similarity between the most similar doc and the input search sentence: ',similarity)
print(similarity_dictionary)
```
